```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def P(n):
    if n == 1:
        return 1
    if n == 2:
        return 1/2
    return 1/2 * P(n - 1) + 1/2 * P(n - 2)

for n in range(1, 26):
    print(f'n={n} P(n)={P(n)}')

def simulate(n, num_samples):
    count = 0  # Count how many times we land on n

    for _ in range(num_samples):
        step = 1
        for i in range(n - 1):  # n - 1 throws
            move = random.choice([1, 2])  # Coin throw
            step += move
            if step == n:
                count += 1
                break

    return count / num_samples # Estimate probability


print("\nApproximate probability of reaching step 25:")
for i in [1000, 10000, 100000]:
    print(f'n={i} trials P(n)={simulate(25, i)}')
```

```python
# Extension

# Write new function P(n, p)
def P_weighted(n, p):
    if n == 1:
        return 1
    if n == 2:
        return p
    return p * P_weighted(n - 1, p) + (1-p) * P_weighted(n - 2, p)

# Plot P(n) against p
x = np.linspace(0, 1, 10000)
y = P_weighted(25, x)

plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel("Probability of moving 1 step ($p$)")
plt.ylabel(f"$P(25)$")
plt.title(f"$P(25)$ vs $p$")
plt.show()

# Approximate p for which P(25) is a minimum
P_min_indx = np.argmin(y)
p_min = x[P_min_indx]
print(f'Approximate p that minimises P(25) = {p_min:.6f}')
```
